```python
def plot_voxel(frame_path, n_bins=8):
    fr = np.load(frame_path)
    fig = plt.figure(figsize=(20., 10.))
    grid = ImageGrid(fig, 111,  # similar to subplot(111)
                     nrows_ncols=(1, n_bins),  # creates 2x2 grid of axes
                     axes_pad=0.1,  # pad between axes in inch.
    )
    for i in range(0, n_bins):
        grid[i-1].imshow(np.swapaxes(fr[:, :, i], 0, 1))
    plt.title(frame_path.split()[-1])
```

```python
base_dir = '/data/gscarpellini/mpii/events/Seq1/imageSequence/video_6/part_1'
frames = sorted(glob.glob(base_dir + '*'))
#path  ='/home/gianscarpe/dev/data/dhp19/voxel_grid_dataset/movements_per_frame/S1_session_1_mov_1_frame_{}.npy'
for frame in frames[:10]:
    plot_voxel(frame, 8)
```
