```python
import h5py
from matplotlib import pyplot as plt
import numpy as np
import os
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
path  ='/home/gianscarpe/dev/data/dataset/h5_dataset_7500_events/movements_per_frame/S1_session_1_mov_1_frame_{}.npy'
```

```python
for frame in range(1, 30):
    fr = np.load(path.format(frame))
    fig = plt.figure(figsize=(20., 10.))
    grid = ImageGrid(fig, 111,  # similar to subplot(111)
                     nrows_ncols=(1, 4),  # creates 2x2 grid of axes
                     axes_pad=0.1,  # pad between axes in inch.
    )
    frame_dir = f'frame_{frame}'
    os.makedirs(frame_dir, exist_ok=True)
    for i in range(1, fr.shape[-1]+1):
        grid[i-1].imshow(fr[:, :, i-1])
    plt.title(frame_dir)
    plt.savefig(f'{frame_dir}.png')
```
